#### 1. Show the most common words in the article.

```python
##get tokens and token.pos
tokens = []
for token in doc:
    if not token.is_stop and not token.is_punct:
        tokens.append((token.lemma_.lower(), token.pos_))

df = pd.DataFrame(tokens, columns=['word', 'pos'])
```

#### 3. Find a subject/object relationship through the dependency parser in any sentence.

```python
#from video
def pr_tree(word, level):
    if word.is_punct:
        return
    for child in word.lefts:
        pr_tree(child, level+1)
    print('\t'* level + word.text + ' - ' + word.dep_)
    for child in word.rights:
        pr_tree(child, level+1)
```

```python
for sentence in doc.sents:
    pr_tree(sentence.root, 0)
    print('-------------------------------------------')
```

#### 4. Show the most common Entities and their types. & 5. Find Entities and their dependency (hint: entity.root.head)

```python
entities = []
for entity in doc.ents:
    entities.append((entity.text.lower(), entity.label_, entity.root.text, entity.root.head))

df_ent = pd.DataFrame(entities, columns=['entity', 'type', 'root', 'head'])

df_ent['ent_type'] = df_ent['entity'] + '(' + df_ent['type'] + ')'
```

##### Common Entities and types

```python
df_ent['ent_type'].value_counts().head(10)
```

##### Dependencies

```python
df_ent.rename(columns={'head': 'dependency'}).groupby(['entity', 'root', 'dependency']).size().reset_index(name='count').sort_values(by='entity', ascending=True).head(20)
```

#### 6. Find the most similar words in the article

```python
df_freq = df['word'].value_counts().reset_index()
df_freq.columns = ['word', 'frequency']
top = df_freq.head(15)['word'].tolist()
```

```python
##Used ChatGPT to assist with the enumeration formula.
similarities = []
token_doc = {token: nlp(token) for token in top}
for i, token_1 in enumerate(top):
    for token_2 in top[i+1:]:
        if token_1 != token_2:
            doc1 = token_doc[token_1]
            doc2 = token_doc[token_2]
            similarity = doc1.similarity(doc2)
            similarities.append((token_1, token_2, similarity))
##Citation: OpenAI. (2024). ChatGPT. Retrieved from: openai.chatgpt.com
```

```python
df_sim = pd.DataFrame(similarities, columns = ['Word 1', 'Word 2', 'Similarity'])
df_sim.sort_values(by='Similarity', ascending = False).head(20)
```
